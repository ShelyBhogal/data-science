```python
import pandas as pd
import seaborn as sns
import numpy as np
import math
import matplotlib.pyplot as plt

# ipywidgets used to interact with graphs
from ipywidgets import interact

# Increase default figure and font sizes for easier viewing
plt.rcParams['figure.figsize'] = (8, 6)
plt.rcParams['font.size'] = 14
# Preserves padding around subplots while setting limits on axis
plt.rcParams["figure.autolayout"] = True
plt.style.use('fivethirtyeight')

# This makes sure that graphs render in your notebook
%matplotlib inline
```

```python
housing = pd.read_csv('./data/housing.csv')

housing = housing.dropna()

housing.head()
```

```python
housing['median_income'].describe()
```

```python
# Two subplots (2 rows, 1 col)
fontsize = 24
fig, ax = plt.subplots(nrows=2, ncols=1,figsize=(15,10))

ax[0].scatter(housing.longitude, housing.median_income, label='Longitude')
ax[1].scatter(housing.latitude, housing.median_income, label='Latitude')

ax[0].set_xlabel('Longitude\n', fontsize=16)
ax[1].set_xlabel('Latitude', fontsize=16)
ax[0].set_ylabel('Median Income', fontsize=16)
ax[1].set_ylabel('Median Income', fontsize=16)

ax[0].set_title('Distribution of Median Income by Longitude', fontsize=fontsize)
ax[1].set_title('Distribution of Median Income by Latitude', pad=10, fontsize=fontsize)
```

```python
# Generate single plot using Seaborn axes-level function
fig, ax = plt.subplots(1, 1, figsize=(15, 10))

sns.scatterplot(x='median_income', y='median_house_value', hue='ocean_proximity', data=housing, ax=ax)
```
