```python
import pandas as pd
import seaborn as sns
import numpy as np
import math
import matplotlib.pyplot as plt
from datetime import timedelta
from datetime import datetime

from statsmodels.tsa.stattools import adfuller
from sklearn.metrics import mean_squared_error
from statsmodels.tsa.arima.model import ARIMA

%matplotlib inline
```

```python
# aapl.csv is available for multivariate time series analysis
shampoo = pd.read_csv("data/shampoo_sales.csv")

# There is no year value, only day-month - start 2019
shampoo.head()
```

```python
shampoo.dtypes
```

```python
shampoo['Temp_date'] = 0

initial_year = 2019

counter = 0

for indx, row in shampoo.iterrows():
    #print(indx)
    if counter < 12:
        year = initial_year
    elif counter >= 12 and counter < 24:
        year = initial_year + 1
    elif counter >= 24:
        year = initial_year + 2
    shampoo['Temp_date'].iloc[indx] = str(year) + "-" + shampoo['Month'].iloc[indx][2:]
    counter += 1
    
```

```python
shampoo['Month'] = shampoo['Temp_date']
shampoo.drop(columns='Temp_date', inplace=True)
shampoo['Month'] = shampoo['Month'] + "-" + "01"

shampoo.head()
```

```python
shampoo = shampoo.set_index(['Month'])
```

```python
shampoo.head()
```

```python
# Use adfuller() to check if Series is stationary (stationarity means variance and mean stay constant over time)

def ad_test(dataset):
    dftest = adfuller(dataset, autolag='AIC')
    
    print("1. ADF : ", dftest[0])
    print("2. P-Value : ", dftest[1])
    print("3. Num Of Lags : ", dftest[2])
    print("4. Num Of Observations Used For ADF Regression:", dftest[3])
    print("5. Critical Values :")
    for key, val in dftest[4].items():
        print("\t",key, ": ", val)
        
```

```python
# If p < 0.05; Data is stationary
# If p > 0.05; Data is not stationary

ad_test(shampoo.Sales)
```

```python
# p-value is greater than 0.05 so data is not stationary - make sure d=1
```

```python
# ARIMA parameters (1,1,2) gives the lowest AIC score
```

```python
X = shampoo.values

# Create train/test set by 
size = int(len(X) * 0.66)
train, test = X[0:size], X[size:len(X)]
```

```python
# Set up train set as nested array
history = [x for x in train]

# Create empty predictions list
predictions = list()
```

```python
# Fit new ARIMA model with order(1,1,2)
for t in range(len(test)):
    model = ARIMA(history, order=(1, 1, 2))
    model_fit = model.fit()
    output = model_fit.forecast()
    y_pred = output[0]
    predictions.append(y_pred)
    obs = test[t]
    history.append(obs)
    print('predicted=%f, expected=%f' % (y_pred, obs))
```

```python
# Validate model performance
rmse = np.sqrt(mean_squared_error(test, predictions))
print('Test RMSE: %.3f' % rmse)
```

```python
plt.plot(test)
plt.plot(predictions, color='red')
plt.show();
```
