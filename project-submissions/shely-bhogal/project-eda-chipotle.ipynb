### Part 5: Calculate the average number of toppings per burrito.

Note: Let's ignore the `quantity` column to simplify this task.

Hint: Think carefully about the easiest way to count the number of toppings!


```python
file_nested_list.head(10)
```

```python
#dataframe filtered to show only items which contain "Burrito"
burrito_df = file_nested_list[ file_nested_list['item_name'].str.contains("Burrito") ]
burrito_df
```

```python
#count all items in list in 'choice_description'
toppings_total = sum(map(len, burrito_df.choice_description))
toppings_total
```

```python
average_burrito_toppings = toppings_total / burrito_df.shape[0]
round(average_burrito_toppings, 3)
```
